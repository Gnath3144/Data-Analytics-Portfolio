```python
import matplotlib.pyplot as plt

monthly_sales = df.groupby("month")["total_sales"].sum().sort_index()

plt.figure()
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
category_sales = df.groupby("product_category")["total_sales"].sum().sort_values()

plt.figure()
category_sales.plot(kind="bar")
plt.title("Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
df['transaction_hour'] = pd.to_datetime(df['transaction_time'].astype(str), format='%H:%M:%S').dt.hour
hourly_sales = df.groupby('transaction_hour')['total_sales'].sum()

plt.figure()
hourly_sales.plot(kind="bar")
plt.title("Sales by Hour")
plt.xlabel("Hour")
plt.ylabel("Total Sales")
plt.show()
```

```python
category_sales.plot(kind="pie")
plt.title("Category Sales Distribution")
plt.ylabel("")
plt.show()
```
